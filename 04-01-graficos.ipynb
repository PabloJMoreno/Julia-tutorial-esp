# Generando graficos

## Funciones para graficar  

```julia
using Pkg 
Pkg.add("Plots")      # instalando el paquete 'Plots', esto es equivalente a '] add Plots'
```

```julia
using Plots
```

```julia
plot(
    sin,                          # funcion a graficar, seno
    -2π, 2π,                      # rango de valores
    )  
```

```julia
plot(
    x -> sin(x)^2 + cos(x)^3,
    -2π, 2π
    )  
```

```julia
plot(
    x -> -log(x),
    0, 1000, 
    label="-log(x)", 
    xlabel="val", 
    ylabel="-log(x)", 
    title="Behavior of Negative Log"
    )  
# Graficos son por defecto en formato SVG 
```

## Graficando otros valores

```julia
x = [1:10]
```

```julia
y1, y2 = rand(10), rand(10);
```

```julia
plot(x, y1, label="y1")  
plot!(x, y2, label="y2")      # preste atencion a (!)
```

```julia
# una manera más sencilloa de hacerlo

plot(x,   [y1, y2],    label=["Primero" "Segundo"])     

# y puede ser una lista  
# observa que no hay coma (,) en el argumento 'label' 
```

```julia
plot(x,   [y1, y2],  seriestype=:scatter)     
```

```julia
plot(x,   [y1, y2],  seriestype=:scatter, thickness_scaling = 0.7, opacity=0.8, smooth=true)     

# se puede agregar una linea de regresion con smooth=true
```

```julia
y1, y2, y3, y4 = rand(10), rand(10), rand(10), rand(10)

plot(x,   [y1, y2, y3, y4], opacity=0.8, palette = :Dark2_5)

# mas sobre graficado en: https://docs.juliaplots.org/latest/generated/colorschemes/
```
